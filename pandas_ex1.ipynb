```python
import pandas as pd
uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/6faa8b659e98fe76d59431b14ce8cfd5315066e8/aula4.1/movies.csv"

filmes = pd.read_csv(uri) # dataFrame

filmes.head()
```

```python
uriNotas = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/6faa8b659e98fe76d59431b14ce8cfd5315066e8/aula4.1/ratings.csv"

notas = pd.read_csv(uriNotas) #dataFrame
notas.head()
```

```python
notas["rating"] # series
notas["rating"].head() # assim como nos df, posso buscar os 5 primeiros das series
```

```python
notas["rating"].unique() # busco os unicos na series
#ao analisar esses dados, podemos concluir que esse site tem avaliações com "meios" valores.
```

```python
notas["rating"].min() # minima nota, r: 0.5
notas["rating"].mean() # média, r: 3.5015
notas["rating"].max() # nota maxima, r: 5.0
```

```python
filmes.describe() # metodo descrbe descreve um dataFrame com valores estatísticos
```
